**ÇOKLU DOĞRUSAL REGRESYON MODEL TUNİNG**

```python
from sklearn.model_selection import train_test_split
```

```python
X_train ,X_test ,y_train, y_test =train_test_split(X,y,test_size=0.20, random_state=11)
```

```python
X_train.head()
```

```python
lm= LinearRegression()
model= lm.fit(X_train,y_train)
```

```python
#eğitim hatası
np.sqrt(mean_squared_error(y_train,model.predict(X_train)))
```

```python
#test hatası
np.sqrt(mean_squared_error(y_test,model.predict(X_test)))
```

```python
# k katlı cross validation bizim için çok önemli çünkü eğitim için seçeceğimiz yüzde 80 lik kısmı nerden alacağımızı bu yöntem ile buluruz
from sklearn.model_selection import cross_val_score
```

```python
mse=np.mean(-cross_val_score(model, X_train,y_train,cv=10, scoring="neg_mean_squared_error"))
mse
```

```python
rmse= np.sqrt(np.mean(-cross_val_score(model, X_train,y_train,cv=10, scoring="neg_mean_squared_error")))
rmse
```
